```python
import requests
import json
import os
import base64
import time

def get_image_base(file):
    with open(file,'rb') as f:
        content = base64.b64encode(f.read())
        return content 

file_names = os.listdir('images')
for file_name in file_names:
    if 'jpg' in file_name:
        img_path = 'images/' + file_name
        #print(img_path)

        access_token = "24.358b25b13fff49a27fe0041fb2ecbc6e.2592000.1554356031.282335-15684400"

        request_url = 'https://aip.baidubce.com/rest/2.0/face/v3/detect'
        request_url = request_url + '?access_token=' + access_token

        params = {
        'image': get_image_base(img_path),
        'image_type': 'BASE64',
        'face_field': 'age,beauty,gender'
        }

        response = requests.post(request_url,data=params)
        json_result = json.loads(response.text)
        code = json_result['error_code']
        if code == 222202:  #  参数格式错误，过滤掉非人物
            continue

        try:
            gender = json_result['result']['face_list'][0]['gender']['type']
            if gender == 'male':   #过滤掉男性
                continue  
            beauty = json_result['result']['face_list'][0]['beauty']
            new_beauty = round(beauty/10,1)   # 原来的打分是100满分，现改为10分
            print(img_path,new_beauty)

            if new_beauty >= 8:
                os.rename(os.path.join('images',file_name),os.path.join('8分',str(new_beauty)+'+'+file_name))
            elif new_beauty >= 7 and new_beauty < 8:
                os.rename(os.path.join('images',file_name),os.path.join('7分',str(new_beauty)+'+'+file_name))
            elif new_beauty >= 6 and new_beauty < 7:
                os.rename(os.path.join('images',file_name),os.path.join('6分',str(new_beauty)+'+'+file_name))
            elif new_beauty >= 5 and new_beauty < 6:
                os.rename(os.path.join('images',file_name),os.path.join('5分',str(new_beauty)+'+'+file_name))
            else:
                os.rename(os.path.join('images',file_name),os.path.join('其他分',str(new_beauty)+'+'+file_name))
            time.sleep(1)
        except KeyError:
            pass
        except TypeError:
            pass
```
